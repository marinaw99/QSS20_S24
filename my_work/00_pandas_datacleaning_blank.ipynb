## 2. Creating new columns/transforming their type 

### Simple filtering

`np.where` is in the numpy package (aliased as `np`) and operates similar to `ifelse` in R

But there is a built in pandas way to emulate this.

**Task**: create an indicator `is_theft` for any offense that contains the word "THEFT"
    
**Task**: create an indicator `is_theft_notmotor` for any offense that contains the word "THEFT" but does not contain the word "MOTOR"

```python
df["is_theft"] = df.OFFENSE.str.contains("THEFT")
```

```python
# ANS
df["is_theft"] = np.where(df.OFFENSE.str.contains("THEFT"), True, False)
```

```python
df["is_theft_notmotor"] = df.OFFENSE.str.contains("THEFT") & ~df["OFFENSE"].str.contains("MOTOR")
```

```python
# ANS
df["is_theft_notmotor"] = np.where(df.OFFENSE.str.contains("THEFT"), True, False) & np.where(df.OFFENSE.str.contains("MOTOR"), False, True)
```

### np.select

**Task**: create a new variable, `offense_summary`, where you:
        
- Recode theft offenses that use a gun or knife as the method (`METHOD`) as: violent theft
- Recode non-theft offenses that use a gun or knife as the method as: violent other
- Recode all other as non-violent 

```python
df.METHOD.unique()
df.OFFENSE.unique()
```

```python
# Define the categories
cat = ["Violent Theft", "Violent Other"]

# Define the conditions
conditions = [
    df["OFFENSE"].str.contains("THEFT") & df["METHOD"].str.contains("GUN|KNIFE"),
    df["METHOD"].str.contains("GUN|KNIFE") & ~df["OFFENSE"].str.contains("THEFT")
]

# Recode w/ np.select
df["offense_summary"] = np.select(conditions, cat, default = "Non-Violent")
```

```python
# ANS
categories = ["violent theft", "violent other"]
conditions = [(df['is_theft']== True) & (df['METHOD'].str.contains("GUN", "KNIFE")),
              (df['is_theft'] == False) & (df['METHOD'].str.contains("GUN", "KNIFE"))]
df["offense_summary"] = np.select( conditions, categories, default = "non-violent")
print(df[["METHOD", "is_theft", "offense_summary"]])
```

### map.recode

**Task**: recode shifts that are MIDNIGHT or EVENING as "nighttime"; code other shift to daytime

```python
df.SHIFT.unique()
shifts = ["MIDNIGHT", "EVENING", "DAY"]
period = ["nighttime", "nighttime", "daytime"]

conversion_dict = dict(zip(shifts, period))
df["SHIFT_NEW"] = df.SHIFT.map(conversion_dict)

df.SHIFT_NEW.unique()
```

```python
# ANS
cat = ["nighttime"]
condition = [df.SHIFT.isin(["MIDNIGHT", "EVENING"])]
df["SHIFT_RECODE"] = np.select(condition, cat, default = "daytime")
df.head()
```
